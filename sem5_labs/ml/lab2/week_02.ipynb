```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x = [0, 20, 40, 60, 80]
y=[]
for i in x:
    y.append(i*2)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
fig1 = plt.figure()
ax1= fig1.add_axes([0, 0, 1, 1])
ax2=fig1.add_axes([0.2,0.5,.2,.2])
ax1.plot(x, y,label='ax1 plot')
ax2.plot(x, y,label='ax2 plot')
ax1.legend()
ax2.legend()
plt.show()
```

```python
df=pd.read_csv('/home/ustudent/Desktop/220962049_aiml_a1/lab2/company_sales_data.csv')
```

```python
fig = plt.figure()
x = df['month_number']
y = df['total_profit']
ax= fig.add_axes([0, 0, 1, 1])
plt.xlabel('month number')
plt.ylabel('total profit')
ax.plot(x, y)
plt.show()
```

```python
fig = plt.figure()
x = df['month_number']
y = df['total_profit']
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(
    x, y,
    linestyle='--',       
    color='red',          
    marker='o',           
    markerfacecolor='black',
    linewidth=3           
)
ax.set_xlabel('Month Number')
ax.set_ylabel('Sold units number')  
ax.set_title('Plot of Total Profit by Month') 
ax.legend(['Total Profit'], loc='lower right')
plt.show()
```

```python
columns_to_drop = ['total_profit', 'total_units']
df_pivot = df.drop(columns=columns_to_drop, errors='ignore')
df_pivot = df_pivot.pivot_table(index='month_number')
fig, ax = plt.subplots()
df_pivot.plot(ax=ax, linestyle='-', marker='o',linewidth=3)
ax.set_xlabel('Month')
ax.set_ylabel('Units Sold')
ax.legend(title='Product', loc='best')
plt.show()
```

```python
columns_to_drop.append('month_number')
df_dropped=df.drop(columns=columns_to_drop,errors='ignore')
x=df_dropped.sum()
fig, ax = plt.subplots()
ax.pie(x, labels=x.index, autopct='%1.1f%%', startangle=90)
plt.show()
```
