```python
#q16
seasons = set(df.season)
season_runs = {}
match_runs = {}
for r in range(len(fd)):
    id = fd.match_id[r]
    if id not in match_runs:
        match_runs[id] = fd.total_runs[r]
    else:
        match_runs[id] += fd.total_runs[r]
    
season_to_id = {}
for r in range(len(df)):
    season, id = df.season[r], df.id[r]
    if season not in season_to_id:
        season_to_id[season] = [id]
    else:
        season_to_id[season] += [id]

for year in season_to_id:
    ids = season_to_id[year]
    runs = 0
    for id in ids:
        runs += match_runs[id]
    season_runs[year] = runs
    
season_runs
```
