```python
a=df3.cut.replace('Ideal',0).replace('Premium',1).replace('Very Good',2).replace('Good',3).replace('Fair',4)
df3['ncut']=a
df3.drop(['cut'],axis=1)
b=df3.color.replace('G',0).replace('E',1).replace('F',2).replace('H',3).replace('D',4).replace('I',5).replace('J',6)
df3['ncolor']=b
df3.drop(['color'],axis=1)
c=df3.clarity.replace('SI1',0).replace('VS2',1).replace('SI2',2).replace('VS1',3).replace('VVS2',4).replace('VVS1',5).replace('IF',6).replace('I1',7)
df3['nclarity']=c
df3.drop(['clarity'],axis=1)
df3
```

```python
Var_Corr = df3.corr()

sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)
```

```python
df4=df3.drop(['clarity','cut','color','price','nclarity'],axis=1)
X=df4
y=df3.price
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
model = LinearRegression().fit(X_train, y_train)
print("Linear Regression accuracy: ", model.score(X_test, y_test))
```
